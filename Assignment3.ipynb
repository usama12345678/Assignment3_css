```python
#q3.1
import nltk.data
from nltk import tokenize
with open('Article.txt', 'r') as file:
    data = file.read().replace('Â', '').replace('â€”','').replace('\xa0',' ')
#tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
data2=sorted(set(data.lower().strip().split()))
separated_text=tokenize.sent_tokenize(data)
total_sentence=len(separated_text)
#print(separated_text)
data3=[x.lower() for x in separated_text]
```

```python
#q3.2
word_frequency={separated_text[sentences_keys]:{word:separated_text[sentences_keys].split().count(word) for word in separated_text[sentences_keys].split()} for sentences_keys in range(len(separated_text))}
print(word_frequency)
```

```python
#q3.3 and 3.4
dct_term_frq={separated_text[sentences_keys]:{word:round(separated_text[sentences_keys].split().count(word)/len(separated_text[sentences_keys].split()),3) for word in separated_text[sentences_keys].split()} for sentences_keys in range(len(separated_text))}
```

```python
#3.5
word_per_doc={}
merge_list=[]
for word in data2:
    for sentences_keys in range(len(data3)):
        if word in sorted(set(data3[sentences_keys].split())):
            word = word.strip().replace('"','').replace(',','')

            merge_list.append(word)
            word_per_doc[word]=merge_list.count(word)
print(word_per_doc)
```

```python
#3.6
import math
sent_idf_dict={}
for sent_keys, sent_values in dct_term_frq.items():
    word_idf_dict={}
    for word_key, word_val in sent_values.items():
        up_word_key=word_key.lower().strip().replace('"','').replace(',','')
        word_idf_dict[word_key]=round(math.log10(total_sentence/word_per_doc[up_word_key]),2)
    sent_idf_dict[sent_keys]=word_idf_dict
print(sent_idf_dict)
```

```python
#3.7
tf_idf_sent_dict={}   
for (tfsent_keys, tfsent_values), (idfsent_keys, idfsent_values) in zip(dct_term_frq.items(),sent_idf_dict.items()):
    tf_idf_word_dict={}
    for (tfword_keys, tfword_values),(idfword_keys, idfword_values) in zip(tfsent_values.items(), idfsent_values.items()):
        tf_idf_word_dict[tfword_keys]= round(float(tfword_values * idfword_values),2)
    tf_idf_sent_dict[tfsent_keys] = tf_idf_word_dict
print(tf_idf_sent_dict)
```

```python
#3.8
sent_weight={}
for weight_sent_keys, weight_sent_val in tf_idf_sent_dict.items():
    sentence_weight=0
    word_len=len(weight_sent_val)
    for weight_word_keys, weight_word_val in weight_sent_val.items():
        sentence_weight=sentence_weight+weight_word_val
    sent_weight[weight_sent_keys]= round(sentence_weight/word_len, 2)
print(sent_weight)
```

```python
#3.9 Average Sentence Weight
sum_of_sent=0
for val in sent_weight:
    sum_of_sent = sum_of_sent + sent_weight[val]
print()
threshold= sum_of_sent/len(sent_weight)
```

```python
#3.10 Summary
total_sentences=0
summary=''
for sent in separated_text:
    if sent_weight[sent]>= threshold:
        summary +=" "+ sent
        total_sentences +=1
print(summary)
print(total_sentences)
```
